```python
import numpy as np
roll_number = 102316066
np.random.seed(roll_number)
```

```python
sales_data = np.random.randint(1000, 5000, size=(12, 4))
sales_data
```

```python
import pandas as pd

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df_sales = pd.DataFrame(sales_data, columns=categories, index=months)
df_sales
```

```python
df_sales.head()
```

```python
df_sales.describe()
```

```python
total_sales_category = df_sales.sum()
total_sales_month = df_sales.sum(axis=1)

total_sales_category, total_sales_month
```

```python
sales_growth = df_sales.pct_change() * 100
sales_growth
```

```python
df_sales['Total Sales'] = df_sales.sum(axis=1)
df_sales['Growth Rate'] = df_sales['Total Sales'].pct_change() * 100
df_sales
```

```python
if roll_number % 2 == 0:
    df_sales['Electronics'] *= 0.9
else:
    df_sales['Clothing'] *= 0.85

df_sales
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for category in categories:
    plt.plot(df_sales.index, df_sales[category], label=category)

plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_sales[categories])
plt.title('Sales Distribution by Category')
plt.ylabel('Sales Units')
plt.xticks(rotation=45)
plt.show()
```

```python
def swap_elements(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return lst

lst = [10, 20, 30, 40, 50]
swap_elements(lst, 1, 3)
```

```python
def swap_elements_in_set(s, index1, index2):
    lst = list(s)
    lst[index1], lst[index2] = lst[index2], lst[index1]
    return set(lst)

s = {10, 20, 30, 40, 50}
swap_elements_in_set(s, 1, 3)
```
